# 1. First-Order Markov Process Simulation

A first-order Markov chain satisfies the Markov property:

$$
P(X_{n+1} = j \mid X_n = i) = P_{ij}
$$

We simulate a sequence of states using:
- A transition probability matrix **P**
- A starting state **s₀**
- Number of steps **N**


```python
def simulate_markov_first_order(P, s0, N):
    states = list(range(P.shape[0]))
    sequence = [s0]

    for _ in range(N):
        current = sequence[-1]
        next_state = random.choices(states, weights=P[current])[0]
        sequence.append(next_state)

    return sequence

# Example transition matrix
P = np.array([
    [0.7, 0.3],
    [0.4, 0.6]
])

sequence = simulate_markov_first_order(P, s0=0, N=20)
sequence
```

# 2. n-Step Transition Probabilities

For a Markov chain with transition matrix \( P \):

$$
P^{(n)} = P^n
$$

This gives the probability of going from state *i* to state *j* in *n* steps.


```python
def n_step_transition(P, n):
    return np.linalg.matrix_power(P, n)

n = 5
P_5 = n_step_transition(P, n)
P_5
```

# 3. Steady-State Distribution

The steady-state vector \( \pi \) satisfies:

$$
\pi P = \pi
$$

with the normalization condition:

$$
\sum_{i} \pi_i = 1
$$

This represents the long-run behavior of the chain.


```python
def steady_state(P, max_iter=1000, tol=1e-8):
    n = P.shape[0]
    pi = np.ones(n) / n  # initial guess

    for _ in range(max_iter):
        new_pi = pi @ P
        if np.linalg.norm(new_pi - pi) < tol:
            break
        pi = new_pi

    return pi

steady_state_vector = steady_state(P)
steady_state_vector
```

# 4. Second-Order (Higher-Order) Markov Process

A second-order chain satisfies:

$$
P(X_{n+1} = k \mid X_n = j, X_{n-1} = i)
$$

Instead of a 2D matrix, transitions require a 3D tensor:

- **P[i][j][k] = probability of going to k after (i → j)**


```python
def simulate_markov_second_order(P2, s0, s1, N):
    sequence = [s0, s1]
    num_states = P2.shape[0]

    for _ in range(N):
        i, j = sequence[-2], sequence[-1]
        next_state = random.choices(range(num_states), weights=P2[i][j])[0]
        sequence.append(next_state)

    return sequence

# Example second-order transition tensor (2 states)
P2 = np.array([
    [[0.6, 0.4], [0.5, 0.5]],
    [[0.3, 0.7], [0.2, 0.8]]
])

simulate_markov_second_order(P2, s0=0, s1=1, N=15)
```
