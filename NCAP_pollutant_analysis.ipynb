```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("city_hour.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df_gurugram = df[df['City'] == 'Gurugram']
display(df_gurugram.tail(40))
```

```python
print(f"Original DataFrame shape: {df.shape}")
print(f"Filtered DataFrame (Gurugram only) shape: {df_gurugram.shape}")
```

```python
df_gurugram.info()
```

```python
df_gurugram = df_gurugram.drop(columns=['Xylene', 'NH3'])
df_gurugram.info()
```

```python
df_gurugram['Datetime'] = pd.to_datetime(df_gurugram['Datetime'])
df_gurugram.info()
```

```python
columns_to_impute = ['PM2.5', 'PM10', 'NO', 'NO2', 'NOx', 'CO', 'SO2', 'O3', 'Benzene', 'Toluene']
df_gurugram[columns_to_impute] = df_gurugram[columns_to_impute].interpolate(method='linear')
df_gurugram.info()
```

```python
print('Remaining missing values after linear interpolation:')
print(df_gurugram[columns_to_impute].isnull().sum())
```

```python
columns_with_remaining_nans = ['PM2.5', 'PM10', 'NO', 'SO2', 'Benzene', 'Toluene']
df_gurugram[columns_with_remaining_nans] = df_gurugram[columns_with_remaining_nans].bfill().ffill()
print('Remaining missing values after bfill then ffill:')
print(df_gurugram[columns_to_impute].isnull().sum())
```

```python
print(f"Data type of 'Datetime' column: {df_gurugram['Datetime'].dtype}")
```

```python
df_gurugram.head(10)
```

```python
df_gurugram['AQI'] = df_gurugram['AQI'].interpolate(method='linear').bfill().ffill()
```

```python
def get_aqi_bucket(aqi):
    if pd.isna(aqi):
        return np.nan
    elif aqi <= 50:
        return 'Good'
    elif aqi <= 100:
        return 'Satisfactory'
    elif aqi <= 200:
        return 'Moderate'
    elif aqi <= 300:
        return 'Poor'
    elif aqi <= 400:
        return 'Very Poor'
    else:
        return 'Severe'

df_gurugram['AQI_Bucket'] = df_gurugram.apply(lambda row: get_aqi_bucket(row['AQI']) if pd.isna(row['AQI_Bucket']) else row['AQI_Bucket'], axis=1)
```

```python
print('Missing values after handling AQI and AQI_Bucket:')
print(df_gurugram[['AQI', 'AQI_Bucket']].isnull().sum())
```

```python
display(df_gurugram.head())
```

```python
df_gurugram['Hour'] = df_gurugram['Datetime'].dt.hour
display(df_gurugram.head())
```

```python
hourly_aqi = df_gurugram.groupby('Hour')['AQI'].mean().reset_index()
display(hourly_aqi.head(20))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.lineplot(x='Hour', y='AQI', data=hourly_aqi)
plt.title('Average Hourly AQI in Gurugram')
plt.xlabel('Hour of Day')
plt.ylabel('Average AQI')
plt.xticks(range(0, 24))
plt.grid(True)
plt.show()
```

```python
def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Summer'
    elif month in [6, 7, 8, 9]:
        return 'Monsoon'
    elif month in [10, 11]:
        return 'Post-Monsoon/Autumn'
    else:
        return 'Unknown'

df_gurugram['Season'] = df_gurugram['Datetime'].dt.month.apply(get_season)
display(df_gurugram.head())
```

```python
seasonal_hourly_aqi = df_gurugram.groupby(['Season', 'Hour'])['AQI'].mean().reset_index()
display(seasonal_hourly_aqi.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

seasons = seasonal_hourly_aqi['Season'].unique()

plt.figure(figsize=(15, 12))

for i, season in enumerate(seasons):
    plt.subplot(2, 2, i + 1)  # Create a 2x2 grid of subplots
    season_data = seasonal_hourly_aqi[seasonal_hourly_aqi['Season'] == season]
    # Add markers and set a distinct color for each line in the subplot
    sns.lineplot(x='Hour', y='AQI', data=season_data, marker='o', color=sns.color_palette('deep')[i])
    plt.title(f'Average Hourly AQI in Gurugram - {season}', fontsize=14)
    plt.xlabel('Hour of Day', fontsize=12)
    plt.ylabel('Average AQI', fontsize=12)
    plt.xticks(range(0, 24), fontsize=10)
    plt.yticks(fontsize=10)
    plt.grid(True, linestyle='--', alpha=0.6)

plt.suptitle('Seasonal Hourly AQI Variation in Gurugram', fontsize=16, y=1.02)
plt.tight_layout()
plt.show()
```

```python
pollutant_columns = ['PM2.5', 'PM10', 'NO', 'NO2', 'NOx', 'CO', 'O3', 'Benzene', 'Toluene']
print(f"Identified pollutant columns: {pollutant_columns}")
```

```python
hourly_pollutant_avg = df_gurugram.groupby('Hour')[pollutant_columns].mean().reset_index()
display(hourly_pollutant_avg.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(15, 8))
for pollutant in pollutant_columns:
    sns.lineplot(x='Hour', y=pollutant, data=hourly_pollutant_avg, label=pollutant)

plt.title('Average Hourly Variation of Major Pollutants in Gurugram', fontsize=16)
plt.xlabel('Hour of Day', fontsize=12)
plt.ylabel('Average Concentration', fontsize=12)
plt.xticks(range(0, 24))
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(title='Pollutant', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
hourly_aqi_pm = df_gurugram.groupby('Hour')[['PM2.5', 'PM10', 'AQI']].mean().reset_index()
display(hourly_aqi_pm.head())
```

```python
def get_breathability_category(aqi):
    if aqi <= 50:
        return 'Good'
    elif aqi <= 100:
        return 'Moderate'
    elif aqi <= 200:
        return 'Poor'
    else:
        return 'Very Poor'

aqi_thresholds = {
    'Good': 50,
    'Moderate': 100,
    'Poor': 200,
    'Very Poor': hourly_aqi_pm['AQI'].max() + 50 # Set a high value for the upper bound of Very Poor for visualization purposes
}

aqi_colors = {
    'Good': 'green',
    'Moderate': 'yellow',
    'Poor': 'orange',
    'Very Poor': 'red'
}

print("Breathability category function and visualization thresholds/colors defined.")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(15, 8))

# Plot horizontal AQI category bands
current_y = 0
for category, threshold in aqi_thresholds.items():
    plt.axhspan(current_y, threshold, color=aqi_colors[category], alpha=0.1, label=f'AQI {category}')
    current_y = threshold

# Plot the average hourly AQI
sns.lineplot(x='Hour', y='AQI', data=hourly_aqi_pm, marker='o', color='blue', linewidth=2, label='Average AQI')

# Add vertical shaded regions for specific time blocks
plt.axvspan(5, 8, color='grey', alpha=0.15, label='Early Morning (5-8 AM)')
plt.axvspan(12, 16, color='purple', alpha=0.15, label='Afternoon (12-4 PM)')
plt.axvspan(21, 24, color='black', alpha=0.15, label='Night (9 PM-12 AM)')
plt.axvspan(0, 5, color='black', alpha=0.15) # Night (12 AM-5 AM)

plt.title('Average Hourly AQI with Breathability Categories and Time Blocks in Gurugram', fontsize=16)
plt.xlabel('Hour of Day', fontsize=12)
plt.ylabel('Average AQI', fontsize=12)
plt.xticks(range(0, 24), fontsize=10)
plt.yticks(fontsize=10)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(title='Legend', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(15, 8))
sns.lineplot(x='Hour', y='PM2.5', data=hourly_aqi_pm, label='PM2.5', marker='o')
sns.lineplot(x='Hour', y='PM10', data=hourly_aqi_pm, label='PM10', marker='o')

plt.title('Average Hourly PM2.5 and PM10 Concentration in Gurugram', fontsize=16)
plt.xlabel('Hour of Day', fontsize=12)
plt.ylabel('Average Concentration', fontsize=12)
plt.xticks(range(0, 24))
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(title='Pollutant')
plt.tight_layout()
plt.show()
```

```python
df_gurugram['DayOfWeek'] = df_gurugram['Datetime'].dt.day_name()

day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
df_gurugram['DayOfWeek'] = pd.Categorical(df_gurugram['DayOfWeek'], categories=day_order, ordered=True)

day_hour_aqi = df_gurugram.groupby(['DayOfWeek', 'Hour'], observed=False)['AQI'].mean().unstack()
display(day_hour_aqi.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate the correlation matrix for the pollutant columns
correlation_matrix = df_gurugram[pollutant_columns].corr()

# Create the heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Heatmap of Major Pollutants in Gurugram', fontsize=16)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.yticks(rotation=0, fontsize=10)
plt.tight_layout()
plt.show()
```
